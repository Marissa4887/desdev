# Collection
The data collection will use Twitter from a stream using the hashtag #ITLLBEOKAY

```python
CONFIG = configparser.ConfigParser()
CONFIG.read('credentials.ini')
```

```python
class StdOutListener(tweepy.Stream):
    def on_status(self, status):
        # saves the tweet object
        user = status.author.screen_name
        location = status.user.location
        timeTweet = status.created_at
        source = status.source
        tweet = status.text
        writer.writerow([tweet,timeTweet,user,location,source])
```

```python
def start_streaming():
     # Opening a CSV file to save the gathered tweets
        with open("new_tweets2.csv", 'w', encoding="utf-8", errors="ignore") as file:
            global writer
            writer = csv.writer(file)

            # Add a header row to the CSV
            writer.writerow(["Tweet", "Date", "User","Location","Source"])

            # Creating a Stream
            try:
                listener = StdOutListener(CONFIG['DEFAULT']['consumer_key'], CONFIG['DEFAULT']['consumer_secret'], CONFIG['DEFAULT']['access_token'], CONFIG['DEFAULT']['access_token_secret'])
                listener.filter(track=["#ITLLBEOKAY"])  
            except KeyboardInterrupt:
                #User pressed ctrl+c or cmd+c -- get ready to exit the program
                print("- KeyboardInterrupt caught. Closing stream and exiting.")
                listener.disconnect()
                pass 
```

```python
start_streaming()
```

## Analysis -1

Since this is an exploration of the collected data, I need to use the methods to understand what story the data is telling. The brief read of the data, cleaning of the data, and using stopwards to remove redundant or useless data (in a repeating fashion). This should result in a clearer understanding of the data and prepare it for sentiment analysis to test my assumption of the data. 

```python
df_tweets.head()
```

```python
re_list = ['(https?://)?(www\.)?(\w+\.)?(\w+)(\.\w+)(/.+)?', '@[A-Za-z0-9_]+','#']
combined_re = re.compile( '|'.join( re_list) )
regex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
```

```python

token = WordPunctTokenizer()
def cleaning_tweets(t):
    del_amp = BeautifulSoup(t, 'lxml')
    del_amp_text = del_amp.get_text()
    del_link_mentions = re.sub(combined_re, '', del_amp_text)
    del_emoticons = re.sub(regex_pattern, '', del_link_mentions)
    lower_case = del_emoticons.lower()
    words = token.tokenize(lower_case)
    result_words = [x for x in words if len(x) > 2]
    return (" ".join(result_words)).strip()
```

```python
cleaned_tweets = []
for i in range(0,len(df_tweets['Tweet'])):
    cleaned_tweets.append(cleaning_tweets((df_tweets.Tweet[i])))
print(cleaned_tweets[0:50])
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["will", "let", "keep", "way","today","friday", "time for newmusicfriday what are you listening today","newmusicfriday","wanderertola","wanderedtola","aqui"])
```

# Analysis - 2
Looking at the sentiment and testing several words of interest, I found most were positive or neutral. The word I used, again, was love for the latest analysis. It did score highly as a 1.0 on positive and 0.64 on compound polarity score. And while this seems good a review of the top tweets, shows that those tweets above 0.5 all seem like bots (like the @ShawnMLoveBot). And the first "real" tweet has a sentiment score of 0.45.

```python
# Next, we initialize VADER so we can use it within our Python script
sid = SentimentIntensityAnalyzer()
```

```python
sid.polarity_scores("love")
```

```python
def calculate_sentiment(text):
    # Run VADER on the text
    scores = sid.polarity_scores(text)
    # Extract the compound score
    compound_score = scores['compound']
    # Return compound score
    return compound_score
```

```python
calculate_sentiment("love")
```

```python
df_tweets['Sentiment Score'] = df_tweets['Tweet'].apply(calculate_sentiment)
df_tweets.sort_values(by='Sentiment Score', ascending=False)[:15]
```
