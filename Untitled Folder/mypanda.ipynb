# Stage Three: Clean data
I replaced missing values with a filler(NA) in three columns, replace shorthand from the dataset with readable language, and checked and removed duplicates.

```python
# There were blanks in the rating, gender and race categories so I filled with NA. 

toon_df['rating'] = toon_df['rating'].fillna('NA')
toon_df['gender'] = toon_df['gender'].fillna('NA')
toon_df['race'] = toon_df['race'].fillna('NA')


toon_df
```

```python
# Clarified the header character and added character_name, then renamed disability as well. 

toon_df.rename(columns={"character": "character_name", "disability": "differently_abled"})
```

```python
# I did a remove duplicate check on this database. I first checked and got a count (length), then removed any and rechecked 
# the count. They are the same so no duplicates in this database.

print(toon_df.duplicated())
```

```python
toon_df.drop_duplicates(inplace = True)
print(toon_df.duplicated())
```

# Stage Four: Data Plot


```python
# Here I plotted the genders by year. The year is a string value 
toon_df['year'].value_counts()[:16].plot(kind="barh", title="All LGBTQ Characters")
```

```python
# The plot was good but I needed the actual counts to really see how many entries by each year?

toon_df['year'].value_counts()
```

```python
# Grouping by Year the Race and gender helped me visualize the data within the race cagtegory with the understading that
# the database was all LGBTQ characters. But the actual counts were better than what I could plot at this point.

toon_df.groupby(['year', 'race'])['gender'].count().plot(kind="barh", title=" Gender and Race by Year")
```

```python
# Now just a look at the data tells me a better story.

toon_df.groupby(['year', 'race'])['gender'].count()
```

```python
# Now just looking at .

toon_df.groupby(['orientation', 'gender'])['platform'].count().plot(kind="pie", title="Orientation and Gender by Plaform")
```

# Stage Five: Comparisons and Claims


The data collected here from 1996 to 2021 represents 16 genders and 12 orientations which includes some unknown or uncategorized characters. It is assumed that all in this database are within the LGBTQ spectrum of representation, either implicitly or explicitly, in childrenâ€™s TV. In this exploration I investigated if any of these numbers were in response to some outside events in LGBTQ history and what is being represented on childrenâ€™s TV.

Reviewing genders by race and year, the chart showed (although the graph was difficult to read) many people of color (POC) being represented in 2018 which was also a year in which there were more lead actors and TV news POC according to Statista.com (Stoll, 2021). Further there was a peak in 2004 and 2004 of representation which was also the year that same-sex marriage became legal in Canada. Also, there was a significant increase in representation in 2014 which was when same-sex marriage was being debated in the Supreme Court of the United States and preceding the year in which it was to be law.

In this exploration I investigated the orientation and gender carried by platforms. The pie chart above shows the most concentrated representation is Gay (Gay, Man and Gay, Cis, Man) and Lesbian (Lesbian, Cis, Woman, and Lesbian, Nonbinary, Lesbian, Woman) characters (represented as the largest slices of the pie accumulatively). This demonstrates just how much Gays and Lesbians are more, perhaps overly, represented than the other orientations on childrenâ€™s TV.

In summary, the networks (aka platforms) seem to reflect the nation's major LGBTQ events and decisions to provide more POC representation within the TV networks. They also seem to focus on characters that are mainly Gay and Lesbian.
